```python
import pandas as pd
```

```python
df17=pd.read_csv('cybercrime17.csv',encoding='unicode_escape')
df18=pd.read_csv('cybercrime18.csv',encoding='unicode_escape')
df19=pd.read_csv('cybercrime19.csv',encoding='unicode_escape')
df20=pd.read_csv('cybercrime20.csv',encoding='unicode_escape')
df21=pd.read_csv('cybercrime21.csv',encoding='unicode_escape')
```

```python
df17.drop(columns=['S. No (Col.1)','Category (Col.2)'],inplace=True)
df18.drop(columns=['S. No (Col.1)','Category (Col.2)'],inplace=True)
df19.drop(columns=['S. No'],inplace=True)
df20.drop(columns=['Si. No. (Col. 1)'],inplace=True)
df21.drop(columns=['Sl. No. Col. (1)'],inplace=True)
```

```python
dfs=[df17,df18,df19,df20,df21]
def add_year(dfs):
    for i in range(2017,2022):
        for df in dfs:
            df['year']=i;
add_year(dfs)
```

```python
dict_tables={ 
    "table1": df17,
    "table2": df18,
    "table3": df19,
    "table4": df20,
    "table5":df21
}

def transform_columns(dict_tables):
    for table in dict_tables.values():
        table.columns=["crime_head",
        "pending_prev_year",
        "reported_during_year",
        "reopened_for_investigation",
        "total_for_investigation",
        "not_investigated_157_1_b",
        "transferred_other_state_or_agency",
        "withdrawn_by_govt",
        "fr_non_cognizable",
        "fr_false",
        "fr_other_13",
        "fr_other_14",
        "fr_other_15",
        "fr_total",
        "chargesheet_prev_year",
        "chargesheet_during_year",
        "chargesheet_total",
        "disposed_by_police",
        "quashed_investigation",
        "stayed_investigation",
        "pending_end_year",
        "chargesheeting_rate_pct",
        "pendency_pct","year"]
transform_columns(dict_tables)
```

```python
df_final = pd.concat(dfs, ignore_index=True)
df_final
```

```python
df_final.to_csv('cybercrime_data_cleaned.csv',index=False,encoding='utf-8')
```
