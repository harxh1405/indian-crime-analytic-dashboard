```python
import pandas as pd
```

```python
url='https://artefacts.data.gov.in/economic-and-productive-aspects-of-the-prison-inmates/'
list_of_tables=pd.read_html(url,header=0)

df = list_of_tables[3]

def transform_common_column(tables):
    for df in tables:
        df.rename(columns={df.columns[0]: "State/UT"}, inplace=True)

transform_common_column(list_of_tables[:-1])
```

```python
from functools import reduce

df_merged = reduce(
    lambda left, right: pd.merge(left, right, on="State/UT", how="inner"),
    list_of_tables[:-2]
)
df_merged.to_csv("cleaned_prison_inmate_data.csv",index=False, encoding="utf-8")
```
