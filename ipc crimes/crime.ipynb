```python
import pandas as pd
```

```python
df16=pd.read_csv('crime16.csv')
df17=pd.read_csv('crime17.csv')
df18=pd.read_csv('crime18.csv')
df19=pd.read_csv('crime19.csv')
df20=pd.read_csv('crime20.csv')
df21=pd.read_csv('crime21.csv')
```

```python
df17=df17.drop(columns=['Category (Col.2)','S. No (Col.1)']).reset_index(drop=True)
df17['year']=2017
df17.head(2)
# df17.head(2)
# df17.shape
```

```python
df18=df18.drop(columns=['Category (Col.2)','S. No (Col.1)']).reset_index(drop=True)
df18['year']=2018

# df18.drop('Category (Col.2)',axis=1,inplace=True)
df18.head(2)
# df18.shape
```

```python
df19=df19.drop(columns=['S. No']).reset_index(drop=True)
df19['year']=2019

df19.head(2)
# df19.shape
```

```python
df20=df20.drop(columns=['Si. No. (Col. 1)']).reset_index(drop=True)
df20['year']=2020

df20.head(2)
# df20.shape
```

```python
df21=df21.drop(columns=['Sl. No. Col. (1)']).reset_index(drop=True)
df21['year']=2021

df21.head(2)
# df21.shape
```

```python
dict_tables={ 
    "table1": df17,
    "table2": df18,
    "table3": df19,
    "table4": df20,
    "table5":df21
}

def transform_columns(dict_tables):
    for table in dict_tables.values():
        table.columns=["crime_head",
        "pending_prev_year",
        "reported_during_year",
        "reopened_for_investigation",
        "total_for_investigation",
        "not_investigated_157_1_b",
        "transferred_other_state_or_agency",
        "withdrawn_by_govt",
        "fr_non_cognizable",
        "fr_false",
        "fr_other_13",
        "fr_other_14",
        "fr_other_15",
        "fr_total",
        "chargesheet_prev_year",
        "chargesheet_during_year",
        "chargesheet_total",
        "disposed_by_police",
        "quashed_investigation",
        "stayed_investigation",
        "pending_end_year",
        "chargesheeting_rate_pct",
        "pendency_pct","year"]

transform_columns(dict_tables)
```

```python
dfs = [df17, df18, df19, df20, df21]
df_final = pd.concat(dfs, ignore_index=True)
df_final



# df_final = pd.concat(
#     dfs, 
#     keys=[2017, 2018, 2019, 2020,2021]
# ).reset_index(level=0).rename(columns={"level_0": "year"})
```
